Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np
#crea datos con beta
a = 5
b = 3
n=1000
vals = np.random.beta(a=a, b=b, size=n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=100)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad exp con lamda=1

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

# exp(1)
x= np.arange(1,10,0.001)
plt.plot(x, expon.pdf(x,a))
plt.title('Función de densidad exp(1)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

QQ plot usando statsmodels. 

```python
import statsmodels.api as sm
import scipy.stats as stats
a=1
expon_dist = stats.expon(scale=a)

fig = sm.qqplot(vals, dist=expon_dist, line='45')

plt.title("Exp teorica con lamda=1")
plt.show()
```

QQ plot contra exp(0.1)

```python
import statsmodels.api as sm
import scipy.stats as stats
expon_dist = stats.expon(scale=0.1)

# Generar el QQ plot comparando los datos 'vals' con la distribución exponencial personalizada
fig = sm.qqplot(vals, dist=expon_dist, line='45')

plt.title("Exp teorica con lamda=0.1")
plt.show()
```
